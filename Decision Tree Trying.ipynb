```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.feature_names)
print(iris.target_names)
print(iris.data[0])
print(iris.target[0])
```

```python
import numpy as np
from sklearn import tree
test_idx = [0,50,100]

train_target = np.delete(iris.target, test_idx)
train_data = np.delete(iris.data, test_idx, axis=0)

# testing data, for testing data we are only focused on test_idx

test_target = iris.target[test_idx]
test_data = iris.data[test_idx]
```

## Classifier

```python
# Creating decision Tree
clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_target)
```

```python
pred_result = clf.predict(test_data)
pred_accuracy = np.mean(np.equal(pred_result, test_target))
print("Hasil prediksi :", pred_result)
print("True target :", test_target)
print(" Prediksi Akurasi :", pred_accuracy)
```

```python
# Visualisasi Pohon (Membuat dalam bentuk pdf)
import os
import graphviz
os.environ["PATH"] += os.pathsep + 'C:/Program Files/Graphviz/bin/'

dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, 
                                filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("iris")
```
